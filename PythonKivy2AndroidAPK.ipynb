from kivy.app import App
from kivy.uix.widget import Widget
from kivy.properties import NumericProperty, ReferenceListProperty, ObjectProperty
from kivy.vector import Vector
from kivy.clock import Clock
from kivy.core.window import Window

# Ball class
class Ball(Widget):
    velocity_x = NumericProperty(0)
    velocity_y = NumericProperty(0)
    velocity = ReferenceListProperty(velocity_x, velocity_y)

    def move(self):
        self.pos = Vector(*self.velocity) + self.pos

# Paddle class
class Paddle(Widget):
    def move(self, new_x):
        self.x = new_x
        # Keep paddle inside screen
        if self.x < 0:
            self.x = 0
        if self.x > Window.width - self.width:
            self.x = Window.width - self.width

# Game main class
class PaddleBounceGame(Widget):
    ball = ObjectProperty(None)
    player = ObjectProperty(None)
    score = NumericProperty(0)

    def serve_ball(self):
        self.ball.center = self.center
        self.ball.velocity = Vector(4, 4)

    def update(self, dt):
        self.ball.move()

        # Bounce off walls (left, right, top)
        if self.ball.y < 0:
            self.ball.velocity_y *= -1
        if self.ball.x < 0 or self.ball.x > Window.width - self.ball.width:
            self.ball.velocity_x *= -1
        if self.ball.y > Window.height - self.ball.height:
            self.ball.velocity_y *= -1

        # Bounce off paddle
        if self.ball.collide_widget(self.player):
            self.ball.velocity_y *= -1
            self.score += 1

        # Game over condition
        if self.ball.y < 0:
            self.serve_ball()
            self.score = 0

    def on_touch_move(self, touch):
        self.player.move(touch.x)

# App class
class PaddleBounceApp(App):
    def build(self):
        game = PaddleBounceGame()
        game.serve_ball()
        Clock.schedule_interval(game.update, 1.0 / 60.0)
        return game

if __name__ == '__main__':
    PaddleBounceApp().run()
